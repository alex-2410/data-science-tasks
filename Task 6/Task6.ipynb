```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

## Importing the dataset

```python
iris = datasets.load_iris()
dataset = pd.DataFrame(iris.data, columns = iris.feature_names)
dataset.head()
```

```python
y = iris.target
```

## Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(dataset, y)
```

## Visualizing the decision tree

```python
from sklearn import tree

plt.figure(figsize = (25, 20))
tree.plot_tree(classifier, feature_names = iris.feature_names, class_names = iris.target_names, filled = True)
plt.show()
```

Utility function to display the class

```python
def print_class(x):
  switcher = {
    0: 'Iris-setosa',
    1: 'Iris-versicolour',
    2: 'Iris-virginica',
  }

  return switcher.get(x[0], None)
```

## Making a prediction

```python
prediction = classifier.predict(np.array([5.1, 3.5, 1.4, 4.2]).reshape(1, -1))
print(print_class(prediction))
```
